```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
figures_file_loc = '../reports/figures/'
```

```python
data = [['1 hr ago', ('block', 'datetime'), 41.353744],
        ['mean',('NA'), 32.299322], 
        ['gbdt-simple', ('block', 'datetime'), 30.724769]]

df = pd.DataFrame(data, columns = ['Name', 'Parameters', 'SmapeError']) 
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
sns.barplot(data=df, x='Name', y='SmapeError', ax=ax)

ax.set_ylabel('SMAPE Error')

plt.savefig(figures_file_loc+'simple_models_smape_error.png')
```

```python
data = [['mean',('NA'), 32.299322], 
        ['ARIMA\n+Differencing', ('block', 'datetime'), 29.557592],
        ['ARIMA\n-Differencing', ('block', 'datetime'), 25.155208]]

df_arima = pd.DataFrame(data, columns = ['Name', 'Parameters', 'SmapeError']) 
```

```python
fig, ax = plt.subplots()
sns.barplot(data=df.append(df_arima), x='Name', y='SmapeError', ax=ax)

ax.set_ylabel('SMAPE Error')
plt.xticks(rotation=50)
plt.tight_layout()
plt.savefig(figures_file_loc+'arima_models_smape_error.png')
```

```python
data = [['mean\n1-block',('NA'), 16.720730027418398], 
        ['ARIMA\n1-block', ('block', 'datetime'), 12.588608600918798]]

df_1block = pd.DataFrame(data, columns = ['Name', 'Parameters', 'SmapeError']) 
```

```python
fig, ax = plt.subplots()
sns.barplot(data=df.append(df_arima).append(df_1block), x='Name', y='SmapeError', ax=ax)

ax.set_ylabel('SMAPE Error')
plt.xticks(rotation=50)
plt.tight_layout()
plt.savefig(figures_file_loc+'arima_1block_models_smape_error.png')
```

```python
data = [['ARIMA\n1-block\n+RF', ('block', 'datetime', 'weather'), 11.226885375524542],]
df_rf = pd.DataFrame(data, columns = ['Name', 'Parameters', 'SmapeError']) 
```

```python
fig, ax = plt.subplots()
sns.barplot(data=df.append(df_arima).append(df_1block).append(df_rf), x='Name', y='SmapeError', ax=ax)

ax.set_ylabel('SMAPE Error')
plt.xticks(rotation=50)
plt.tight_layout()
plt.savefig(figures_file_loc+'arima_1block_rf_models_smape_error.png')
```
